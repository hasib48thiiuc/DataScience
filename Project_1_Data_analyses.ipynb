```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/final_dataset_after_data_cleaning.csv')
```

#Descriptive_statistics:

```python
descriptive_stats = df.describe()
```

```python
print(descriptive_stats)
```

#Univariate Analysis

```python
df.head()
```

```python
# removing column
#df.drop(columns='Unnamed: 0', axis=1, inplace=True)

string_columns = []

for column in df.columns:
  if df[column].dtype == 'object':
    string_columns.append(column)

df = df.drop(string_columns, axis=1)
```

```python
data_dict = df.to_dict(orient='record')
print(data_dict)

#print(type(data_dict['Material']))
```

```python
print(data_dict[0])
print(type(data_dict[0]))
for key in data_dict[0]:
  print(key)
  print(type(key))
```

```python
for key in data_dict[0]:
  plt.hist(df[key], bins=10)
  plt.xlabel(key)
  plt.ylabel('Frequency')
  plt.title(f'Histogram of {key}')
  plt.show()
```

# Bivariate Analysis

```python
for key in data_dict[0]:
  for i in data_dict[0]:
    plt.scatter(df[key], df[i])
    plt.xlabel(key)
    plt.ylabel(i)
    plt.title(f'Scatter Plot of {key} vs. {i}')
    plt.show()
```

# Correlation Analysis:

```python
correlation_matrix = df.corr()
```

```python
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
